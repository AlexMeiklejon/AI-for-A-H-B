## 1.2 Material and Cultural Context

To better understand the textile samples within the catalogue, it is important to consider the materials and cultural contexts from which they originate. ### Bark Cloth & Tapa

Bark cloth, often referred to as tapa, is a non-woven textile made by soaking and beating the inner bark of trees such as paper mulberry, breadfruit, or fig until the fibres fuse into a flexible fabric (Kamen-Kaye, 1984). Unlike woven textiles, its structure comes from this felting process, where strips can be layered, joined, and extended to create larger or thicker sheets (Kamen-Kaye, 1984). 
Across the Pacific, it exists under different names — such as ngatu in Tonga, masi in Fiji, and siapo in Samoa — and varies in texture, thickness, and decoration depending on local materials and techniques (Harris, 2020). Designs are often applied using natural dyes and tools, ranging from geometric patterns to motifs inspired by plants, animals, and mythology, reflecting regional variation and cultural knowledge (Lennard and Mills, 2020). Historically, bark cloth was used for a wide range of purposes including clothing, bedding, wall hangings, and ceremonial exchange, making it a highly versatile material within everyday and ritual life (MacLean, 2015). Despite the introduction of European fabrics, it continues to be produced in parts of the Pacific today, maintaining its importance as both a practical material and a cultural artefact.

### Social Use

Tapa cloth in the Pacific served both everyday and ceremonial functions, reflecting broader social structures. While it was commonly used for basic clothing such as loincloths, it also held significant ritual value, being reserved for ceremonies, festivals, and the presentation to honoured guests (Neich and Pendergrast, 1997). Its production further reflected gendered roles, with tapa-making often described as a key form of women’s creative expression, while in some regions men produced tapa specifically for ritual objects (Neich and Pendergrast, 1997). Beyond its practical use, tapa also communicated social identity, as patterns could indicate clan allegiance or regional origin (Alsford, 2025). As a result, tapa functioned not only as a material resource but as a medium through which status, identity, and cultural belonging were expressed. It could also play a role in social exchange, for example being presented to honoured guests, reinforcing its importance within systems of value and reciprocity (Neich and Pendergrast, 1997).

### Locations

The textile samples in the catalogue originate from various islands across the Pacific, including Otaheite (Tahiti), Owyhee (Hawaii), and the Sandwich Islands. These locations reflect the wide geographical distribution of tapa production across Polynesia, where bark cloth was made and used in diverse cultural contexts. Although the material itself was broadly shared across the region, local variations in production, design, and use demonstrate the distinct identities of different island communities (Neich and Pendergrast, 1997).

However, the place names used within the catalogue also reflect a European perspective shaped by colonial exploration. Terms such as “Sandwich Islands” and “Owyhee” were used by Captain James Cook and other European explorers, rather than the Indigenous names used by local populations. For example, Hawaii was first recorded in European maps in 1778 when Cook identified and named the archipelago as the Sandwich Islands (Newbolt, 1929). This naming system illustrates how Pacific spaces were documented and redefined through European frameworks, often obscuring Indigenous understandings of place.

As a result, the locations recorded in the catalogue do not simply indicate geographic origin, but also reveal how knowledge about the Pacific was constructed and organised through colonial encounters. 

### Colonial Context 

The bark cloth samples within this catalogue must also be understood within the context of European exploration and colonial collecting practices. During Captain Cook’s voyages, Pacific textiles were actively collected by European explorers, who viewed them as objects of scientific and cultural interest rather than as items embedded within living traditions (Thomas, 2018). These materials were often removed from their original contexts and transformed into “specimens,” arranged in sample books that emphasised their material and aesthetic qualities over their cultural meanings. As a result, objects such as tapa cloth became part of European systems of knowledge production, where they were categorised, studied, and displayed across museums and collections. While this process contributed to wider awareness of Pacific material culture, it also involved the fragmentation of objects that originally functioned as complete and socially embedded textiles. From a postcolonial perspective, this reduction of bark cloth into detached samples can be understood as a form of cultural simplification, stripping the material of its full social and ceremonial significance (Thomas, 2018). Consequently, the catalogue itself reflects not only the textiles it contains, but also the historical conditions under which they were collected, interpreted, and preserved. 


### Cultural Context

Beyond its practical and decorative uses, tapa production was often closely tied to community life, identity, and tradition. The process of making bark cloth could function as a communal and culturally significant activity, with knowledge and techniques passed down through generations (Alsford, 2025). In many contexts, tapa also held symbolic value, being used in rituals, ceremonies, and exchanges that reinforced social relationships and cultural continuity (Neich and Pendergrast, 1997). Designs and materials could carry meaning, reflecting connections to ancestry, environment, and local belief systems. However, much of this cultural significance is absent within the catalogue itself, where the textiles are primarily described in terms of their material properties, origins, and uses. This highlights a key limitation of the archive: while it preserves the physical qualities of the textiles, it often excludes the cultural knowledge and perspectives of the communities who produced them. As a result, the dataset constructed from the catalogue reflects both what is recorded and what is missing, emphasising the importance of contextual research in interpreting these materials.

## 2. Design Approach
## 2.1 Method

The overall process can be understood as a sequence of human–AI interaction:

Human: digitises and interprets the physical archive
Dataset: structured into a CSV file for analysis
AI tools: used to analyse patterns and generate outputs
Human: evaluates and contextualises the results
Output: presented as a structured and interpretable engagement with the archive

This process is not linear but iterative, with human decisions shaping how AI is used, and AI outputs informing further interpretation.

## 2.2 Human-AI Collaboration

This project adopts a human–AI collaborative approach, where AI is used to augment rather than replace human interpretation. The dataset was constructed through manual digitisation and interpretation of the textile catalogue, requiring subjective judgement in matching samples to descriptions and interpreting ambiguous entries. AI tools were then applied to analyse and extend this structured data, identifying patterns and generating outputs that would be difficult to observe manually. However, these outputs still require human interpretation to provide cultural and historical context. This reflects a collaborative workflow in which human decision-making and AI-generated insights inform one another, producing a more meaningful engagement with the archive than either could achieve independently.

Generative AI is particularly relevant to this project as it enables the creation of new outputs based on existing data, rather than simply analysing it. While traditional approaches to archival work focus on description and preservation, generative AI introduces the possibility of reinterpretation and transformation. In this context, AI is not used to recreate the archive, but to extend it—highlighting patterns, relationships, and alternative ways of engaging with the material that may not be immediately visible through manual analysis alone.

Digitising the textile catalogue into a structured dataset significantly expands its accessibility. While the original object is limited to a physical archive, the dataset allows the material to be searched, compared, and analysed digitally. When combined with AI tools, this further broadens the potential audience by enabling new forms of interaction, including data-driven insights and generated outputs. This shift from physical object to digital dataset therefore not only preserves the archive but repositions it as a resource that can be engaged with in more dynamic and accessible ways.

This project made use of generative AI tools to support different stages of the workflow. ## 3. Implementation
## 3.1 Building the Dataset

This project involved transforming a physical archival object into a structured dataset. The book contains textile samples attached to its pages, each marked with a small handwritten number. These numbers correspond to descriptions listed elsewhere in the catalogue.

To build the dataset, I first photographed the textile samples and the corresponding descriptive pages. I then manually matched each numbered sample to its written description. This process required careful interpretation, as some entries used abbreviated language such as “ditto,” requiring cross-referencing with previous entries.

The extracted information was then organised into a CSV file. A CSV (Comma-Separated Values) file is a simple format used to store structured data in a table, where each row represents an entry and each column represents a specific attribute. It was appropriate for this project because it allows the textile samples to be organised into clear, consistent categories (such as ID, material, colour, origin, and use), making the data easy to read, analyse, and process. Its simplicity and compatibility with tools like Python also make it well suited for transforming archival information into a structured dataset. Where details were missing or unclear, informed approximations were made based on the visual characteristics of the samples.

This process converted the archive into a structured format that allows the textile specimens to be compared and analysed systematically.

The dataset was loaded into Python using pandas, allowing the CSV file to be read and displayed as a structured table. This makes it possible to explore and analyse the data programmatically.

```python
df['Origin'].value_counts()
import matplotlib.pyplot as plt

# Chart 1: Distribution of samples by origin
origin_counts = df['Origin'].value_counts()

fig, ax = plt.subplots(figsize=(8, 5))
origin_counts.plot(kind='barh', ax=ax, color='#8B6F47', edgecolor='black')
ax.set_title('Textile Samples by Recorded Origin', fontsize=13, pad=15)
ax.set_xlabel('Number of samples')
ax.set_ylabel('Origin (as recorded in catalogue)')
ax.invert_yaxis()  # largest bar at top
plt.tight_layout()
plt.show()
```

The distribution of textile samples by origin shows that the majority of items in the dataset come from the Sandwich Islands, reflecting the strong presence of Hawaiian bark cloth within the catalogue. Other samples originate from Owyhee (Hawaii) and Otaheite (Tahiti), while some entries are listed as unknown. This highlights both the geographical focus of the collection and the limitations of the archival record, where not all samples are clearly documented.

```python
# Classify each sample by broad social/functional use
# This is an interpretive mapping based on the catalogue's descriptions
use_categories = {
    12: 'Ceremonial / Religious',
    5: 'Everyday / Practical',
    6: 'Everyday / Practical',
    19: 'Everyday / Practical',
    20: 'Everyday / Practical',
    22: 'Elite / High-quality',
    23: 'Ceremonial / Religious',
    30: 'Everyday / Practical',
    24: 'Gendered dress',
    25: 'Gendered dress',
    26: 'Decorative / Ornamental',
}
df['Social Use'] = df['ID'].map(use_categories)

# Chart 3: Distribution by social use
use_counts = df['Social Use'].value_counts()

fig, ax = plt.subplots(figsize=(8, 5))
use_counts.plot(kind='barh', ax=ax, color='#6B4423', edgecolor='black')
ax.set_title('Textile Samples by Social / Functional Use', fontsize=13, pad=15)
ax.set_xlabel('Number of samples')
ax.set_ylabel('Social use (interpreted)')
ax.invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Chart 2: Distribution of samples by recorded colour
colour_counts = df['Colour'].value_counts()

# Approximate hex codes that roughly match each described colour
colour_map = {
    'Light beige': '#E8D5B7',
    'Yellow': '#E8C547',
    'Reddish-brown': '#8B4513',
    'Light beige with red speckling': '#D4B896',
    'Reddish-brown striped': '#A0522D',
    'Dark reddish-brown': '#5D2E0D',
    'Brown': '#6F4E37',
}
bar_colours = [colour_map.get(c, '#CCCCCC') for c in colour_counts.index]

fig, ax = plt.subplots(figsize=(8, 5))
colour_counts.plot(kind='barh', ax=ax, color=bar_colours, edgecolor='black')
ax.set_title('Textile Samples by Recorded Colour', fontsize=13, pad=15)
ax.set_xlabel('Number of samples')
ax.set_ylabel('Colour (as recorded in catalogue)')
ax.invert_yaxis()
plt.tight_layout()
plt.show()
```

This grouping is an interpretive step, not a neutral reading of the data. The catalogue's descriptions are brief and sometimes ambiguous, so assigning each sample to a category required judgement. Making this step visible is itself important: it demonstrates how the apparent objectivity of a dataset is shaped by human decisions at every stage of its construction.

This shows that lighter tones dominate the dataset, reflecting the natural appearance of bark cloth and the limited use of dyes in some samples.

## 3.2 Mapping the Archive

Once the dataset was created, several patterns became visible across the textile samples. One of the clearest patterns is the variation in social use. Some textiles are associated with specific groups, such as priests and young women, while others are described as being used by common people or for everyday purposes like bedding. This suggests that textiles were not only functional but also reflected social roles and status.

There is also a clear geographical spread across the dataset, with samples originating from locations such as Sandwich Island, Otaheite (Tahiti), and Owyhee (Hawaii). However, these place names reflect colonial terminology, showing that the archive is already shaped by external perspectives before being transformed into data.

In terms of function, the textiles serve a range of purposes, including clothing, bedding, and decoration. Some are used for practical purposes, while others are linked to ornamentation or ceremonial contexts. This highlights how the same material, bark cloth, could be adapted for different uses.

The dataset also shows variation in production techniques. References to dyeing, oiling, and surface texturing suggest different methods of manufacturing, although these processes are only briefly described.

Overall, structuring the archive into a dataset makes it easier to compare samples and identify patterns. However, it also simplifies the material, as much of the cultural context and meaning behind the textiles is reduced to short descriptive fields.


## 4. AI Limitations

Despite its potential, the use of AI in this project is subject to important limitations. AI systems rely on patterns learned from data and do not possess contextual understanding of cultural or historical significance. As a result, outputs can be incomplete, misleading, or lacking nuance. This reinforces the importance of human oversight, particularly when working with cultural heritage material. The dataset itself also reflects limitations of the original archive, where certain contextual details are absent or unclear, meaning that both the data and the AI outputs must be interpreted critically rather than taken at face value.

A further limitation of this project is the relatively small sample size of the dataset. While the structured data allows for basic pattern identification, the scale is not sufficient for more advanced or meaningful AI-driven analysis. Generative AI methods are typically more effective when applied to larger datasets, where patterns are more robust and diverse. This project should be understood as a proof of concept, demonstrating how archival material can be transformed into data.
