## 1.1 Chosen Object

This project focuses on *A Catalogue of Different Specimens of Cloth collected in the Three Voyages of Captain Cook*. The object is a bound volume containing physical textile samples collected during Cook’s expeditions, with fragments of cloth attached directly to its pages.

The catalogue presents these textiles as “specimens”, reflecting a scientific approach to material culture. However, while the physical qualities of the fabrics are preserved, much of their original cultural context such as their use, meaning, and the communities who produced them is absent.

The object therefore raises important questions about how cultural artefacts are collected, represented, and understood within archival systems.
