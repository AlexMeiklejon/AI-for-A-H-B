
## 1.2 Material and Cultural Context

To better understand the textile samples within the catalogue, it is important to consider the materials and cultural contexts from which they originate. Alongside constructing the dataset from the archive, I conducted additional research to deepen my understanding of the textiles and the contexts in which they were used. This background knowledge provides essential context for interpreting the dataset.

### Bark Cloth & Tapa

Bark cloth, often referred to as tapa, is a non-woven textile made by soaking and beating the inner bark of trees such as paper mulberry, breadfruit, or fig until the fibres fuse into a flexible fabric (Kamen-Kaye, 1984). Unlike woven textiles, its structure comes from this felting process, where strips can be layered, joined, and extended to create larger or thicker sheets (Kamen-Kaye, 1984). 
Across the Pacific, it exists under different names — such as ngatu in Tonga, masi in Fiji, and siapo in Samoa — and varies in texture, thickness, and decoration depending on local materials and techniques (Harris, 2020). Designs are often applied using natural dyes and tools, ranging from geometric patterns to motifs inspired by plants, animals, and mythology, reflecting regional variation and cultural knowledge (Lennard and Mills, 2020). Historically, bark cloth was used for a wide range of purposes including clothing, bedding, wall hangings, and ceremonial exchange, making it a highly versatile material within everyday and ritual life (MacLean, 2015). Despite the introduction of European fabrics, it continues to be produced in parts of the Pacific today, maintaining its importance as both a practical material and a cultural artefact.

### Social Use

Tapa cloth in the Pacific served both everyday and ceremonial functions, reflecting broader social structures. While it was commonly used for basic clothing such as loincloths, it also held significant ritual value, being reserved for ceremonies, festivals, and the presentation to honoured guests (Neich and Pendergrast, 1997). Its production further reflected gendered roles, with tapa-making often described as a key form of women’s creative expression, while in some regions men produced tapa specifically for ritual objects (Neich and Pendergrast, 1997). Beyond its practical use, tapa also communicated social identity, as patterns could indicate clan allegiance or regional origin (Alsford, 2025). As a result, tapa functioned not only as a material resource but as a medium through which status, identity, and cultural belonging were expressed. It could also play a role in social exchange, for example being presented to honoured guests, reinforcing its importance within systems of value and reciprocity (Neich and Pendergrast, 1997).

### Locations

The textile samples in the catalogue originate from various islands across the Pacific, including Otaheite (Tahiti), Owyhee (Hawaii), and the Sandwich Islands. These locations reflect the wide geographical distribution of tapa production across Polynesia, where bark cloth was made and used in diverse cultural contexts. Although the material itself was broadly shared across the region, local variations in production, design, and use demonstrate the distinct identities of different island communities (Neich and Pendergrast, 1997).

However, the place names used within the catalogue also reflect a European perspective shaped by colonial exploration. Terms such as “Sandwich Islands” and “Owyhee” were used by Captain James Cook and other European explorers, rather than the Indigenous names used by local populations. For example, Hawaii was first recorded in European maps in 1778 when Cook identified and named the archipelago as the Sandwich Islands (Newbolt, 1929). This naming system illustrates how Pacific spaces were documented and redefined through European frameworks, often obscuring Indigenous understandings of place.

As a result, the locations recorded in the catalogue do not simply indicate geographic origin, but also reveal how knowledge about the Pacific was constructed and organised through colonial encounters. 

### Colonial Context 

The bark cloth samples within this catalogue must also be understood within the context of European exploration and colonial collecting practices. During Captain Cook’s voyages, Pacific textiles were actively collected by European explorers, who viewed them as objects of scientific and cultural interest rather than as items embedded within living traditions (Thomas, 2018). These materials were often removed from their original contexts and transformed into “specimens,” arranged in sample books that emphasised their material and aesthetic qualities over their cultural meanings. As a result, objects such as tapa cloth became part of European systems of knowledge production, where they were categorised, studied, and displayed across museums and collections. While this process contributed to wider awareness of Pacific material culture, it also involved the fragmentation of objects that originally functioned as complete and socially embedded textiles. From a postcolonial perspective, this reduction of bark cloth into detached samples can be understood as a form of cultural simplification, stripping the material of its full social and ceremonial significance (Thomas, 2018). Consequently, the catalogue itself reflects not only the textiles it contains, but also the historical conditions under which they were collected, interpreted, and preserved. 


### Cultural Context

Beyond its practical and decorative uses, tapa production was often closely tied to community life, identity, and tradition. The process of making bark cloth could function as a communal and culturally significant activity, with knowledge and techniques passed down through generations (Alsford, 2025). In many contexts, tapa also held symbolic value, being used in rituals, ceremonies, and exchanges that reinforced social relationships and cultural continuity (Neich and Pendergrast, 1997). Designs and materials could carry meaning, reflecting connections to ancestry, environment, and local belief systems. However, much of this cultural significance is absent within the catalogue itself, where the textiles are primarily described in terms of their material properties, origins, and uses. This highlights a key limitation of the archive: while it preserves the physical qualities of the textiles, it often excludes the cultural knowledge and perspectives of the communities who produced them. As a result, the dataset constructed from the catalogue reflects both what is recorded and what is missing, emphasising the importance of contextual research in interpreting these materials.

Structuring the catalogue as data makes patterns easier to see, but it also reduces complex objects to short descriptive fields. By making these processes visible alongside the textiles themselves, the project invites the audience to think about how our understanding of Pacific material culture has been shaped at every stage.

## 2. Design Approach
## 2.1 Method

The project follows a workflow built around repeated collaboration between human judgement and AI tools. The process moves through five stages.

**Digitisation.** I photographed the textile samples and their corresponding descriptions in the catalogue, transcribing each entry by hand. This stage required careful interpretation, particularly where entries used abbreviated language ("ditto"), unclear handwriting, or outdated terminology. ChatGPT was used as a support tool here to help interpret these inconsistencies.

**Structuring.** The transcribed information was organised into a CSV dataset, with each sample assigned attributes for ID, description, material, colour, origin, and notes. This stage transformed the catalogue from a physical object into a format that could be searched, compared, and analysed.

**Analysis.** The dataset was loaded into Python using pandas and explored through descriptive statistics and visualisations, identifying patterns across origin, colour, and social use.

**Generation.** Using Microsoft Copilot, I produced AI-generated images that attempted to re-situate selected textiles within the cultural contexts the catalogue does not record. Each prompt was developed using the cultural and material context established in Section 1.2, and outputs were treated as material for critical examination rather than as authoritative reconstructions.

**Interpretation.** Each AI output was reflected on critically — both for what it added to the archival record and for what it revealed about the AI's own framings. These reflections feed back into how the project understands the catalogue as a whole.

Human decisions shape how the AI is used at every step, and the AI's outputs in turn shape how the archival material is understood.

## 2.2 Human-AI Collaboration

This project adopts a human–AI collaborative approach, where AI tools are used to extend human interpretation rather than replace it. The dataset was constructed through manual digitisation and interpretation of the textile catalogue, requiring judgement in matching samples to descriptions and interpreting unclear entries. AI tools were then used to support different stages of the workflow, from cleaning the transcribed data, to writing simple Python code for analysis, to generating images that attempt to re-situate the textiles within the cultural contexts the catalogue does not record. 

Generative AI is particularly relevant to this project because it allows the archive to be extended rather than only described. Traditional archival work focuses on preservation and cataloguing, whereas generative AI introduces the possibility of reinterpretation, by producing new outputs based on existing material that can prompt different ways of engaging with it. Digitising the catalogue as a structured dataset already expands its accessibility, turning a physical object into something that can be searched, compared, and analysed. Combining this with generative AI broadens the potential audience further by enabling new forms of interaction.

Two AI tools were used during the project. ChatGPT (OpenAI) was used to help interpret the catalogue's archival language. The original entries contain inconsistent spelling, abbreviation ("ditto"), and outdated or unclear wording, all of which made direct transcription difficult. ChatGPT was used to suggest standardised modern equivalents for these unfamiliar terms, with each suggestion checked against the original catalogue before being accepted into the dataset. This kept human judgement at the centre of the transcription process while allowing it to move more efficiently than manual interpretation alone.

Microsoft Copilot (using OpenAI's DALL·E 3 model) was used for two purposes. First, it supported the simple Python code used to load and analyse the CSV dataset, generating and suggesting code in real time which kept the technical side of the project accessible without requiring advanced programming knowledge. Second, and more substantially, Copilot was used to generate the AI images discussed in Section 3.3. Each image was produced from a research-informed prompt grounded in the cultural and material context established in Section 1.2, and each output was treated as material for critical examination rather than as an authoritative reconstruction of Pacific cultural practice. 

Both tools were chosen for their accessibility and free availability, which made them practical for a project focused on demonstrating human–AI collaboration rather than technical optimisation.

Across all stages, the use of these tools was iterative. AI suggestions informed the next stage of the work, and the work in turn shaped the prompts and choices guiding the AI. The aim was not to generate final interpretations independently, but to use AI to support the organisation, analysis, and presentation of the material — with human judgement remaining the central element of the process.

## 3. Implementation
## 3.1 Building the Dataset

This project involved transforming a physical archival object into a structured dataset. The book contains textile samples attached to its pages, each marked with a small handwritten number. These numbers correspond to descriptions listed elsewhere in the catalogue.

To build the dataset, I first photographed the textile samples and the corresponding descriptive pages. I then manually matched each numbered sample to its written description. This process required careful interpretation, as some entries used abbreviated language such as “ditto,” requiring cross-referencing with previous entries.

The extracted information was then organised into a CSV file. A CSV (Comma-Separated Values) file is a simple format used to store structured data in a table, where each row represents an entry and each column represents a specific attribute. It was appropriate for this project because it allows the textile samples to be organised into clear, consistent categories (such as ID, material, colour, origin, and use), making the data easy to read, analyse, and process. Its simplicity and compatibility with tools like Python also make it well suited for transforming archival information into a structured dataset. Where details were missing or unclear, informed approximations were made based on the visual characteristics of the samples.

The dataset was loaded into Python using pandas, allowing the CSV file to be read and displayed as a structured table. This makes it possible to explore and analyse the data programmatically. This can be seen below:

```python
import pandas as pd

df = pd.read_csv("data/fabrics.csv")
df
```

Once the dataset was created, several patterns became visible across the textile samples. These are explored both descriptively and through the visualisations below, which map the catalogue's recorded origins, colours, and social uses.

```python
!pip install matplotlib
```

```python
df['Origin'].value_counts()
import matplotlib.pyplot as plt

# Chart 1: Distribution of samples by origin
origin_counts = df['Origin'].value_counts()

fig, ax = plt.subplots(figsize=(8, 5))
origin_counts.plot(kind='barh', ax=ax, color='#8B6F47', edgecolor='black')
ax.set_title('Textile Samples by Recorded Origin', fontsize=13, pad=15)
ax.set_xlabel('Number of samples')
ax.set_ylabel('Origin (as recorded in catalogue)')
ax.invert_yaxis()  # largest bar at top
plt.tight_layout()
plt.show()
```

The distribution of textile samples by origin shows that the majority of items in the dataset come from the Sandwich Islands, reflecting the strong presence of Hawaiian bark cloth within the catalogue. Other samples originate from Owyhee (Hawaii) and Otaheite (Tahiti), while some entries are listed as unknown. This highlights both the geographical focus of the collection and the limitations of the archival record, where not all samples are clearly documented.

The chart highlights the dominance of samples from the Sandwich Islands (a colonial-era name for Hawaii), while the "Unknown" category makes visible the gaps in the archival record itself. Visualising these counts foregrounds not only what the catalogue contains, but what it fails to document.

```python
# Classify each sample by broad social/functional use
# This is an interpretive mapping based on the catalogue's descriptions
use_categories = {
    12: 'Ceremonial / Religious',
    5: 'Everyday / Practical',
    6: 'Everyday / Practical',
    19: 'Everyday / Practical',
    20: 'Everyday / Practical',
    22: 'Elite / High-quality',
    23: 'Ceremonial / Religious',
    30: 'Everyday / Practical',
    24: 'Gendered dress',
    25: 'Gendered dress',
    26: 'Decorative / Ornamental',
}
df['Social Use'] = df['ID'].map(use_categories)

# Chart 3: Distribution by social use
use_counts = df['Social Use'].value_counts()

fig, ax = plt.subplots(figsize=(8, 5))
use_counts.plot(kind='barh', ax=ax, color='#6B4423', edgecolor='black')
ax.set_title('Textile Samples by Social / Functional Use', fontsize=13, pad=15)
ax.set_xlabel('Number of samples')
ax.set_ylabel('Social use (interpreted)')
ax.invert_yaxis()
plt.tight_layout()
plt.show()
```

This chart groups the textile samples by their social or functional use, drawing on the descriptions recorded in the catalogue. Everyday and practical uses dominate, with five samples covering bedding, common dress, and general daily wear. Two samples are linked to ceremonial or religious roles, including textiles worn by priests and used in ritual contexts, and two more are associated with gendered dress, specifically garments worn by young women. A single sample is described as finely manufactured, suggesting an elite or high-quality use, and one is recorded as decorative or ornamental.

```python
# Chart 2: Distribution of samples by recorded colour
colour_counts = df['Colour'].value_counts()

# Approximate hex codes that roughly match each described colour
colour_map = {
    'Light beige': '#E8D5B7',
    'Yellow': '#E8C547',
    'Reddish-brown': '#8B4513',
    'Light beige with red speckling': '#D4B896',
    'Reddish-Brown Speckled': '#A0826D',
    'Reddish-brown striped': '#A0522D',
    'Dark reddish-brown': '#5D2E0D',
    'Brown': '#6F4E37',
}
bar_colours = [colour_map.get(c, '#CCCCCC') for c in colour_counts.index]

fig, ax = plt.subplots(figsize=(8, 5))
colour_counts.plot(kind='barh', ax=ax, color=bar_colours, edgecolor='black')
ax.set_title('Textile Samples by Recorded Colour', fontsize=13, pad=15)
ax.set_xlabel('Number of samples')
ax.set_ylabel('Colour (as recorded in catalogue)')
ax.invert_yaxis()
plt.tight_layout()
plt.show()
```


The chart shows that lighter natural tones dominate the dataset, with five samples recorded as light beige. The remaining colours: yellow, reddish-brown, light beige with red speckling, reddish-brown speckled, reddish-brown striped, and brown, each appear only once. This distribution reflects the appearance of undyed bark cloth, which retains the natural pale colour of the inner bark from which it is made, as well as the limited range of pigments and patterning techniques recorded in the catalogue.

## 3.3 Restoring Context: Human-AI Image Generation

Having structured the catalogue into data, this section takes the opposite approach: using generative AI to attempt to re-situate the textiles within the cultural contexts the archive does not record.

Rather than treating the AI's outputs as authoritative reconstructions, they are used as a critical lens, which is a way of making visible both what the catalogue omitted and what the generative model itself fails to know. The AI is asked, where possible, to produce visualisations that genuinely engage an audience with the textiles as culturally situated objects rather than as decontextualised specimens. The collaboration is understood not as the AI doing creative work on the human's behalf, but as a process in which the human's research shapes the prompts. The AI's training data shapes the outputs, and the human then interprets what those outputs reveal about both the archive and the model.

### Methodology

The image generation work was carried out using Microsoft Copilot, which uses OpenAI's DALL·E 3 model. Copilot was selected on the same grounds as the other AI tools in this project, as it is freely accessible, requires no specialist setup, and supports an iterative conversational workflow that allowed prompts to be refined through dialogue. Four samples were chosen from the dataset to span a range of contextual dimensions: the body and garment (sample 25), material and surface treatment (sample 24), production process and tools (sample 20), and functional decoration in environment (sample 26). For each sample, a single research-informed prompt was developed using the cultural and material context established in Section 1.2, with instructions added to discourage stereotyped Pacific imagery such as flower crowns, palm-tree backdrops, and romanticised tropical framing. Where Copilot's content policy refused a prompt, the refusal was documented as a finding rather than worked around. All prompts and outputs, including discarded variations, were saved alongside the final images.

## Synthesis ##

Across the four samples, a clear pattern emerges in how the AI handles Pacific bark cloth. When the cloth is shown on its own — laid out for study, treated with oil, beaten into texture — the AI represents it with reasonable care. The material qualities recorded in the catalogue, including colour, texture, and even functional treatments such as oiling, are picked up and rendered. But as soon as the cloth is asked to interact with a wider context — a wearer, a vessel, or an ornament — the surrounding elements shift. The Western mannequin in sample 25 and the European-style boat in sample 26 are all drawn from familiar visual templates rather than from Pacific cultural practice. The cloth itself remains identifiably Pacific however the world around it does not. This suggests something specific about the AI's training data: isolated images of Pacific textiles, likely from museum and archival collections, appear to be far more represented than images of those textiles in their cultural contexts of use. The result is that the AI, like the original catalogue, ends up presenting the cloth largely separated from the cultural life it was part of, though for different reasons. Together, the four AI Generated images suggest that AI can usefully extend the archive's record of materials, but cannot, on its own, restore the cultural contexts. 

## Ethical Considerations ##

The use of generative AI to produce images of Pacific cultural practice raises ethical questions worth acknowledging directly. The models used in this project are trained on large datasets, the full contents of which are not publicly known. Images of Indigenous peoples, cultural practices, and ceremonial objects have almost certainly been part of that training data, potentially gathered without the knowledge or consent of the communities involved. The AI-generated images are not presented as authentic reconstructions of Pacific life. They are presented as evidence of how contemporary AI systems represent, reshape, or omit that life.

Alongside these broader concerns, ethical considerations also emerged in a more direct way through the use of AI systems. In particular, Copilot refused to generate certain prompts. An initial request relating to sample 25, which described a Polynesian woman wearing the textile in the late eighteenth century, was declined due to concerns around cultural misrepresentation. This response is significant in itself, reflecting how contemporary AI systems are governed by content policies that restrict the depiction of historical Indigenous peoples. As a result, accepted prompts were reformulated to focus on the textile in isolation or as an object on display, rather than being situated on a human figure. 

## 4. AI Limitations

Despite its potential, the use of AI in this project is subject to important limitations. AI systems rely on patterns learned from data and do not possess contextual understanding of cultural or historical significance. As a result, outputs can be incomplete, misleading, or lacking nuance. This reinforces the importance of human oversight, particularly when working with cultural heritage material. The dataset itself also reflects limitations of the original archive, where certain contextual details are absent or unclear, meaning that both the data and the AI outputs must be interpreted critically rather than taken at face value.

A further limitation of this project is the relatively small sample size of the dataset. While the structured data allows for basic pattern identification, the scale is not sufficient for more advanced or meaningful AI-driven analysis. Generative AI methods are typically more effective when applied to larger datasets, where patterns are more robust and diverse. As such, this project should be understood as a proof of concept, demonstrating how archival material can be transformed into data rather than a fully scalable AI application.
